```python
with open('./dt.html', 'r') as f:
    data = f.read()
# data
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
```

```python
links = soup.find_all('a', href=True)

link_dict = {}
results = []
for link in links:
    link_text = link.get_text()
    link_href = link['href'].split('/')[3:]
    # print(f"Link Text: {link_text}, Href: {link_href}")
    link_dict[link_href[-1]] = link_text
    results.append([link_dict.get(j, j) for j in link_href])
# [[link_dict.get(j, j) for j in i] for i in results]
```

```python
import pandas as pd
```

```python
series_list = [pd.Series(row) for row in results]

# 将 Series 列表转换为 DataFrame
df = pd.DataFrame(series_list)
df
```

```python
df.to_excel('./dt.xlsx', index = False)
```
