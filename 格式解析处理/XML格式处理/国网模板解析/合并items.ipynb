```python
new_data = new_data[['itemname_x','key','value_type_x','description_x','application_x','Module','discovery_rule_x','template','units_x','delay','监控项']]
```

```python
new_data.columns = ['itemname','key','value_type','description','application','Module','监控项','discovery_rule','template','units','delay']
new_data.head()
```

```python
new_data = new_data.drop_duplicates(['key','Module']).reset_index(drop = True)
```

```python
def trans(x):
    if x:
        if type(x) == str:
            if x.endswith('m'):
                x = str(int(x.split('m')[0])*60)+'s'
            elif x.endswith('h'):
                x = str(int(x.split('h')[0])*3600)+'s'
            elif x.endswith('s'):
                x = x
        elif type(x) == int:
            x = str(x)+'s'
    return x
```

```python
new_data['units'] = new_data['units'].apply(lambda x : trans(x))
```
