```python
import yaml
import pandas as pd
```

```python
with open('./awesome-prometheus-alerts/_data/rules.yml', 'r') as f:
    datas = yaml.load(f)
```

```python
sublist = []
for index1 in range(0,len(datas['groups'])):
    for index2 in range(len(datas['groups'][index1]['services'])):
        for index3 in range(len(datas['groups'][index1]['services'][index2]['exporters'])):
            if isinstance(datas['groups'][index1]['services'][index2]['exporters'][index3]['rules'], list):
                for i in datas['groups'][index1]['services'][index2]['exporters'][index3]['rules']:
                    subdict = i
                    subdict['servicesname'] = datas['groups'][index1]['services'][index2]['name']
                    subdict['groupsname'] = datas['groups'][index1]['name']
                    if 'for' not in subdict.keys():
                        subdict['for'] = '0m'
                    if 'name' in datas['groups'][index1]['services'][index2]['exporters'][index3].keys():
                        subdict['exportersname'] = datas['groups'][index1]['services'][index2]['exporters'][index3]['name']
                    sublist.append(subdict)
```

```python
len(sublist)
```

```python
newdata = pd.DataFrame(sublist)
newdata
```

```python
newdata.to_excel('prometheus.xlsx', index = False)
```
