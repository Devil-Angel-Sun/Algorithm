## 使用池

```python
import queue
class OpenGeminiPool(queue.Queue):
    def __init__(self, host, port, username, password, maxsize=0):
        queue.Queue.__init__(self, maxsize=maxsize)
        self._maxsize = maxsize
        self.host = host
        self.port = port
        self.username = username
        self.password = password
    
    def create_connection(self):
        return InfluxDBClient(
            host=self.host,
            port=self.port,
            username=self.username,
            password=self.password
        )
    
    def get(self, timeout=None):
        if self.qsize() < self._maxsize or self._maxsize == 0:
            self.put(self.create_connection())
        return queue.Queue.get(self, timeout)
    
    def release(self, client):
        self.put(client)
class openGemini(object):
    def __init__(self, config):
        self._pool = OpenGeminiPool(
            host = config["opengemini"]["host"],
            port = config["opengemini"]["port"],
            username = config["opengemini"]["user"],
            password = config["opengemini"]["password"],
            maxsize = config["opengemini"]["maxsize"]
        )
        
    def read(self, sql, db):
        # 从连接池获取连接
        client = self._pool.get()
        try:
            return client.query(query=sql, epoch="s", database=db)
        finally:
            # 将连接归还到连接池
            self._pool.release(client)
```

```python
opengemini = openGemini(config)
database = 'NzhEMjlBOTk3MzkxRjk5MjQyMzMzQzI4_apm_metric'
query = "SELECT major_collection_time AS value, time FROM 'jvm.gc' WHERE time > 1709087040s AND time < 1709088000s order by time asc;\
    SELECT used AS value, time FROM 'jvm.memory.heap' WHERE time > 1709087040s AND time < 1709088000s order by time asc"
data = opengemini.read(query, database)
```
